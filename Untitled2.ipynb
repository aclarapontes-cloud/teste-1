# **tarefa** **institucional**

questão 1: Dada a lista de temperaturas registradas durante a semana:
temperaturas = [18, 22, 30, 35, 28, 15, 40]
Classifique cada dia como:
• Abaixo de 20 → "Frio"
• Entre 20 e 30 → "Agradável"
• Acima de 30 → "Quente"
Tarefas:
1. Criar uma lista com as classificações
2. Contar quantos dias houve em cada categoria

```python
temperaturas = [18, 22, 30, 35, 28, 15, 40]

classificacoes = []
frio_count = 0
agradavel_count = 0
quente_count = 0

for temp in temperaturas:
    if temp < 20:
        classificacoes.append("Frio")
        frio_count += 1
    elif 20 <= temp <= 30:
        classificacoes.append("Agradável")
        agradavel_count += 1
    else: # temp > 30
        classificacoes.append("Quente")
        quente_count += 1

print(f"Classificações dos dias: {classificacoes}")
print(f"Contagem por categoria:\nFrio: {frio_count} dias\nAgradável: {agradavel_count} dias\nQuente: {quente_count} dias")
```
